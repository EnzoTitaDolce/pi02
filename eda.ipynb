Importamos las librerías comunes que siempre usamos. Si necesitamos otras las importaremos después.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Tenemos dos datasets para trabajar, uno con homicidios en siniestros viales y otro con lesiones en siniestros viales. El primero contiene datos desde 2016 hasta 2021 y el segundo desde 2019 hasta 2021.
Trabajaremos primero con el de homicidios.

Abrimos el archivo excel para crear un dataframe.

Este archivo excel contiene dos hojas, hechos y victimas. Abrimos cada hoja en un dataframe separado y las concatenamos según el id del hecho. De estea forma tendremos información más completa.

```python
#Veamos la primera hoja que tiene los datos sobre los siniestros
homicidios_hechos = pd.read_excel('datasets/homicidios.xlsx', sheet_name='HECHOS')
homicidios_hechos.head()
```

<h1>Exploremos un poco... </h1>
En un primer vistazo observamos que en la columna tipo de calle aparece el valor GRAL. PAZ y otros que dicen AVENIDA. Podríamos pensar que se trata de un error ya que GRAL. PAZ es el nombre de una calle y no un tipo de calle, pero el diccionario de datos utiliza ese valor para referirse, particularmente, a la avenida Gral. Paz en ambos sentidos incluyendo ingresos, egresos, distribuidores y las calles colectoras.

```python
homicidios_hechos.info()
```

```python
#Vamos con las segunda hoja del dataset
homicidios_victimas = pd.read_excel('datasets/homicidios.xlsx', sheet_name='VICTIMAS')
homicidios_victimas
```

Si comparamos la cantidad de registros de los dataframes veremos que el de victimas tiene 21 registros más. Esto puede deberse a que en el mismo hecho pueden haber fallecido más de una persona.
Veremos los registros duplicados según la columna ID_hecho en homicidios_victimas.

```python
duplicados = homicidios_victimas.loc[homicidios_victimas.duplicated(subset='ID_hecho', keep=False)]
duplicados
```

Efectivamente, vemos que hay más de una víctima en el mismo hecho. Al unir los datasets tendremos toda la información en un solo dataframe y será más fácil de manejar.

Los unimos haciendo coincidir la columna ID del dataframe homicidios_hechos con la columna ID_hecho del dataframe homicidios_victimas.

```python
df = pd.merge(homicidios_hechos,homicidios_victimas,left_on='ID',right_on='ID_hecho')
df
```

```python
#Vamos eliminando los dataframes auxiliares que ya no utilicemos para liberar recursos
del(duplicados, homicidios_hechos, homicidios_victimas)
```

Veamos las columnas del dataframe unido

```python
df.columns
```

<li>Podemos empezar a descartar columnas redundantes, como por ejemplo, las que aportan la fecha o el id del hecho, las víctimas, etc.
Son columnas que repiten la misma información en el dataframe.</li>
<li>Por otra parte también podemos descartar las columnas LUGAR_DEL_HECHO porque su información está repetida en la columna Dirección Normalizada.</li>
<li>Descartaremos la columna XY(CABA) porque con las coordenadas en las columnas pos x y pos y ya es suficiente.</li>
<li>Para crear la base de datos nos conviene más tener la dirección separada en 3 columnas que tenerla toda en una sola.</li>
<li>Por último, podemos descartar la columna participantes porque, según el diccionario de datos, es solo una conjunción entre las columnas Victima y Acusado.</li>

```python
#Eliminamos primero las columnas que repetían información en ambos dataframes antes de unirlos.
df.drop(columns=['FECHA_x','ID_hecho', 'FECHA_y', 'AAAA_y', 'MM_y', 'DD_y', 'VICTIMA_y'], inplace=True)
```

```python
df.head()
```

```python
#Aquí buscamos si hay algún registro en el que la columna XY (CABA) aporte información sobre la ubicación y las columnas pos x y pos y estén vacías.
#Y vemos que no hay ningún registro que cumpla con esas condiciones
df[(df['XY (CABA)'] != 'Point (. .)') & ((df['pos x'] == '.') | (df['pos y'] == '.'))]
```

```python
#Ahora buscamos si hay algún registro en el que la columna XY (CABA) esté sin información pero las columnas pos x y pos y estén completas.
df[(df['XY (CABA)'] == 'Point (. .)') & ((df['pos x'] != '.') | (df['pos y'] != '.'))]
```

Con las búsquedas que hicimos vemos que la columna XY (CABA) no aporta información que no se pueda obtener de las otras. Podemos eliminarla.

Según el diccionario de datos la columna participantes es una conjunción de las columnas victima y acusado. Podemos eliminarla también.

```python
df.drop(columns=['XY (CABA)','PARTICIPANTES'], inplace=True)
```

<h1>Analicemos ahora las columnas que nos dan información sobre el lugar del accidente.</h1>

Tengamos en cuenta que para buscar el lugar en un mapa en la web las columnas pos x y pos y (las coordenadas) serán de gran ayuda. Así que a esas columnas las conservaremos.

Separemos las columnas a analizar en otro dataframe para poder operar sin riesgos sobre el dataframe original hasta que tengamos una acción decidida.

```python
columnas_direccion = df[['ID','LUGAR_DEL_HECHO','TIPO_DE_CALLE','Calle','Cruce','Altura','Dirección Normalizada', 'pos x','pos y']]
columnas_direccion.head()
```

Vamos primero con las columnas que tienen la mayor cantidad de valores nulos, es decir, Cruce y Altura.

Si nos guíamos por el diccionario de datos sabremos que el campo Cruce indica las calles en la que se produjo el accidente. O sea que, si el campo cruce tiene un valor no nulo significa que el accidente se produjo en la intersección de las dos calles mencionadas en la columna LUGAR_DEL_HECHO. Por lo tanto no necesitaríamos el dato de la altura para poder ubicar dicho lugar.
El campo altura,en cambio, es necesario para poder ubicar exactamente el lugar el accidente cuando este no se haya producido en una intersección.

El problema surge cuando ambos campos están vacíos. En ese caso solo nos queda valernos de las coordenadas para ubicar el lugar exacto del accidente y, en caso de ser necesario, averiguar el nombre de las calles de la intersección, o la altura o el kilometro en caso de que sea una autopista.

```python
#Son 44 los registros que no tienen información sobre la altura o el cruce. Pero son aún menos los que tampoco tienen información sobre las coordenadas.
#Vamos a centrarnos primero en estos últimos registros.
columnas_direccion[(columnas_direccion['Cruce'].isna()) & (columnas_direccion['Altura'].isna())]
```

```python
#Los registros sin información sobre el cruce ni altura ni coordenadas son pocos (13) podemos analizarlos uno por uno para tratar de conseguir
#la mayor cantidad de información posible.
columnas_direccion[(columnas_direccion['Cruce'].isna()) & (columnas_direccion['Altura'].isna()) & (columnas_direccion['pos x'] == '.') & (columnas_direccion['pos y'] == '.')]
```

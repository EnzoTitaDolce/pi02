Si comparamos la cantidad de registros de los dataframes veremos que el de victimas tiene 21 registros más. Esto puede deberse a que en el mismo hecho pueden haber fallecido más de una persona.

Efectivamente, vemos que hay más de una víctima en el mismo hecho. Al unir los datasets tendremos toda al información en un solo dataframe y será más fácil de manejar.

```python
df = pd.merge(homicidios_hechos,homicidios_victimas,left_on='ID',right_on='ID_hecho')
df
```
